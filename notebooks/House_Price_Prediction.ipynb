```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the dataset California House

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
type(housing)
```

```python
print(housing.feature_names)
```

Preparing the data

```python
dataset = pd.DataFrame(housing.data,columns=housing.feature_names)
type(dataset)
```

```python
dataset.head(5)
```

```python
dataset['Price(CR)']=housing.target
dataset.head(5)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

# EDA

```python
dataset.corr()
```

```python
sns.pairplot(dataset)
# points that are far from others are outlier
```

```python
fig, ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=dataset,ax=ax)
```

Split the data into independant and dependant features

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
X
```

```python
Y
```

Split the data into train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
## normalization of given points
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_norm=scaler.fit_transform(X_train)
X_train_norm
```

```python
fig, ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=X_train_norm,ax=ax)
```

```python
X_test_norm=scaler.transform(X_test)
X_test_norm
```

```python
fig, ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=X_test_norm,ax=ax)
```

# Model training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train_norm,Y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

# Model Prediction

```python
reg_pred = regression.predict(X_test_norm)
reg_pred
```

Calculate the Error and  recidual

```python
residuals = Y_test-reg_pred
residuals
```

Distribution plot of residuals

```python
sns.displot(residuals, kind='kde')
plt.xlabel('Residuals')
plt.title('Distribution of Residuals')
plt.show()
```

Model Performance

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print(mean_squared_error(Y_test,reg_pred))
print(mean_absolute_error(Y_test,reg_pred))
print(r2_score(Y_test,reg_pred))
```

Save the model

```python
import pickle
pickle.dump(regression,open('model.pkl','wb'))
```
